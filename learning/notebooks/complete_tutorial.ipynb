# PART 5: Understanding Z-Scores
---

## 💡 What is a Z-Score?

A z-score tells you **how many standard deviations away from the mean** a value is.

### The Formula:
```
z = (value - mean) / standard_deviation
```

### Example:
Values: `[10, 20, 30, 40, 50]`

1. **Calculate mean**: (10+20+30+40+50)/5 = **30**
2. **Calculate std**: √[(10-30)² + (20-30)² + ... ]/5 = **14.14**
3. **Calculate z-score for 40**: (40 - 30) / 14.14 = **0.71**

### What Does This Mean?
- **z = 0**: Value is exactly at the mean
- **z = +1**: Value is 1 std dev ABOVE mean
- **z = -1**: Value is 1 std dev BELOW mean
- **z = +0.71**: Value is 0.71 std devs above mean (like our example!)

### Why Use Z-Scores?
1. **Removes abundance weighting** from VPCA loadings
2. **Makes all variables comparable** (same scale)
3. **Mean becomes 0, std becomes 1** for all variables

---

````python
# ============================================================
# CALCULATE Z-SCORES MANUALLY (To Understand the Math)
# ============================================================

print("Manual Z-Score Calculation for VPC1")
print("=" * 60)

# STEP 1: Get VPC1 column as a numpy array
# loadings['VPC1'] = get the VPC1 column from loadings DataFrame
# .values = convert from pandas Series to numpy array (list of numbers)
# WHY numpy array? Because we need to do math operations on it
vpc1_original = loadings['VPC1'].values
print(f"Original VPC1 values: {vpc1_original}")
print()

# STEP 2: Calculate the mean (average)
# np.mean() = numpy function that adds all values and divides by count
# Formula: mean = (sum of all values) / (number of values)
vpc1_mean = np.mean(vpc1_original)
print(f"Mean: {vpc1_mean:.4f}")  # .4f = show 4 decimal places

# STEP 3: Calculate standard deviation
# np.std() = numpy function for standard deviation
# Standard deviation measures "how spread out" the values are
# ddof=0 means "population standard deviation" (divide by N)
# ddof=1 means "sample standard deviation" (divide by N-1)
# We use ddof=0 to match SPSS population std
vpc1_std = np.std(vpc1_original, ddof=0)
print(f"Standard Deviation: {vpc1_std:.4f}")
print()

# STEP 4: Apply the z-score formula to EVERY value
# This is "vectorized" - it applies the operation to entire array at once!
# For each value: z = (value - mean) / std
# Example: if value=0.38, mean=0.35, std=0.05
#          then z = (0.38 - 0.35) / 0.05 = 0.60
vpc1_zscore = (vpc1_original - vpc1_mean) / vpc1_std
print(f"Z-scored values: {vpc1_zscore}")
print()

# STEP 5: Verify the z-scores are correct
# Z-scored data should have:
#   - Mean ≈ 0 (approximately zero)
#   - Std ≈ 1 (approximately one)
print("Verification:")
print(f"  Original mean: {vpc1_mean:.6f}")
print(f"  Z-score mean: {np.mean(vpc1_zscore):.6f} (should be ~0)")
print(f"  Original std: {vpc1_std:.6f}")
print(f"  Z-score std: {np.std(vpc1_zscore, ddof=0):.6f} (should be ~1)")

# ============================================================
# WHAT JUST HAPPENED?
# We took the original VPC1 values (different scales)
# and transformed them to z-scores (same scale, mean=0, std=1)
# This makes them comparable to other VPCs!
# ============================================================
````

# PART 6: Create Z-Score Function
---

## 💡 What is a Function?

A function is **reusable code** - write once, use many times!

### Structure:
```python
def function_name(inputs):
    # do something
    return output
```

### Example:
```python
def add_numbers(a, b):
    result = a + b
    return result

answer = add_numbers(5, 3)  # answer = 8
```

We'll create a function to z-score ANY DataFrame!

---

````python
# ============================================================
# DEFINE FUNCTION TO Z-SCORE A DATAFRAME
# ============================================================

# LINE 1: 'def' means "define a function"
# 'zscore_dataframe' is the function name (we can call it anything)
# '(df)' is the input parameter - the DataFrame to z-score
def zscore_dataframe(df):
    """
    Convert each column of a DataFrame to z-scores.
    
    This is called a "docstring" - it explains what the function does.
    It's optional but good practice!
    
    Parameters:
        df: A pandas DataFrame with numeric columns
    
    Returns:
        A new DataFrame with z-scored columns
        (same structure, but values are standardized)
    """
    
    # LINE 2: Create an empty DataFrame to store results
    # pd.DataFrame() creates a new empty DataFrame
    # index=df.index means use the same row labels as input
    result = pd.DataFrame(index=df.index)
    
    # LINE 3: Loop through each column
    # df.columns gives us the list of column names
    # 'for col in ...' means "for each column name"
    for col in df.columns:
        
        # LINE 4: Get the values from this column
        # df[col] gets the column
        # .values converts it to numpy array
        values = df[col].values
        
        # LINE 5: Calculate mean of this column
        mean = np.mean(values)
        
        # LINE 6: Calculate std of this column
        # ddof=0 for population std (matches SPSS)
        std = np.std(values, ddof=0)
        
        # LINE 7-9: Check for zero standard deviation
        # If all values are the same, std = 0
        # We CAN'T divide by zero! (math error)
        # So we handle this special case
        if std == 0:
            # Print a warning
            print(f"  Warning: {col} has zero variance (all values identical)")
            # Set all z-scores to 0 (can't standardize constant values)
            result[col] = 0
        else:
            # LINE 10: Normal case - apply z-score formula
            # z = (value - mean) / std
            # This is vectorized - applies to entire column at once!
            result[col] = (values - mean) / std
    
    # LINE 11: Return the result
    # 'return' means "give this back to whoever called the function"
    return result

# ============================================================
# FUNCTION DEFINED!
# Now we can call it anytime: z_data = zscore_dataframe(data)
# ============================================================

print("✓ Z-score function defined!")
print("  You can now call: zscore_dataframe(your_dataframe)")
````

````python
# ============================================================
# USE THE FUNCTION TO Z-SCORE OUR DATA
# ============================================================

print("Z-Scoring the Data...")
print("=" * 60)

# LINE 1: Z-score the loadings DataFrame
# zscore_dataframe() is the function we just defined
# loadings is the input (our VPCA loadings)
# z_loadings is the output (z-scored version)
z_loadings = zscore_dataframe(loadings)

# LINE 2-3: Display the result
print("\nZ-scored VPCA Loadings:")
# .round(3) shows only 3 decimal places (easier to read)
print(z_loadings.round(3))
print()

# LINE 4: Z-score the spectral library
# Same process - call function with library DataFrame
z_library = zscore_dataframe(library)

# LINE 5-6: Display result
print("Z-scored Spectral Library:")
print(z_library.round(3))

# ============================================================
# NOW WE HAVE:
# - z_loadings: z-scored VPCA loadings (mean=0, std=1)
# - z_library: z-scored spectral library (mean=0, std=1)
# Both are ready for regression!
# ============================================================
````

````python
# ============================================================
# VERIFY Z-SCORING WORKED CORRECTLY
# ============================================================

print("Verification: Checking z-scored loadings")
print("(Each column should have mean ≈ 0, std ≈ 1)")
print("=" * 60)

# Loop through each VPC column
for col in z_loadings.columns:
    # Calculate mean and std of the z-scored column
    # .mean() calculates the average
    # .std(ddof=0) calculates population standard deviation
    mean = z_loadings[col].mean()
    std = z_loadings[col].std(ddof=0)
    
    # Print the results
    # If z-scoring worked:
    #   mean should be very close to 0 (like 0.000001)
    #   std should be very close to 1 (like 0.999999)
    print(f"{col}: mean = {mean:.6f}, std = {std:.6f}")

print()
print("✓ If mean ≈ 0 and std ≈ 1 for all columns, z-scoring worked!")
````

# PART 7: Understanding Linear Regression
---

## 💡 What is Linear Regression?

Regression finds the **relationship** between variables.

### The Model:
```
Y = β₀ + β₁X₁ + β₂X₂ + β₃X₃ + ... + error
```

### In Our Case:
```
VPC1 = β₀ + β₁(Water) + β₂(Chlorophyll) + β₃(Clay) + ... + error
```

### What Each Part Means:
- **Y** = What we want to explain (VPC loadings)
- **X** = What might explain it (spectral library materials)
- **β** = Coefficients (how much each material contributes)
- **β₀** = Intercept (baseline value)

### Example:
```
VPC1 = 0.1 + 0.8(Chlorophyll) + 0.3(Water) - 0.2(Clay)
```
This tells us:
- **Chlorophyll** strongly increases VPC1 (+0.8)
- **Water** moderately increases VPC1 (+0.3)
- **Clay** slightly decreases VPC1 (-0.2)

---

## 🎯 Key Outputs:

### R² (R-squared)
- How much variance is explained (0 to 1)
- **0.0** = explains nothing (0%)
- **0.5** = explains half (50%)
- **1.0** = explains everything (100%)
- **Higher is better!**

### Coefficients (β values)
- **Positive (+)** = material increases with VPC
- **Negative (-)** = material decreases with VPC
- **Larger absolute value** = stronger relationship

### P-values
- Statistical significance
- **p < 0.05** (less than 0.05) = significant (95% confident)
- **p < 0.01** (less than 0.01) = highly significant (99% confident)
- **p > 0.05** (greater than 0.05) = not significant (might be random)

---

````python
# ============================================================
# RUN A SIMPLE REGRESSION FOR VPC1
# ============================================================

print("Simple Regression: VPC1 ~ All Materials")
print("=" * 70)

# STEP 1: Define Y (dependent variable)
# This is what we want to EXPLAIN
# We're trying to explain VPC1 using the materials
y = z_loadings['VPC1']
print(f"Y (dependent variable): VPC1")
print(f"  Shape: {y.shape} (9 values, one per band)")
print()

# STEP 2: Define X (independent variables)
# These are what EXPLAIN Y
# All the materials in the spectral library
X = z_library
print(f"X (independent variables): All materials")
print(f"  Shape: {X.shape} ({X.shape[0]} bands × {X.shape[1]} materials)")
print(f"  Materials: {list(X.columns)}")
print()

# STEP 3: Add constant (intercept) term
# sm.add_constant() adds a column of 1s
# WHY? This represents β₀ (the intercept) in the equation
# Without it, the line is forced through origin (0,0)
X = sm.add_constant(X)
print(f"Added constant (intercept) column")
print(f"  New shape: {X.shape}")
print()

# STEP 4: Create and fit the regression model
# sm.OLS() creates an Ordinary Least Squares model
# OLS = the standard regression method
# .fit() finds the best coefficients
model = sm.OLS(y, X).fit()
print("✓ Model fitted!")
print()

# STEP 5: Display key results
print(f"R-squared: {model.rsquared:.4f}")
print(f"  This means the model explains {model.rsquared*100:.1f}% of variance")
print()
print(f"Adjusted R-squared: {model.rsquared_adj:.4f}")
print(f"  (Adjusted for number of predictors)")

# ============================================================
# WHAT JUST HAPPENED?
# We ran a regression to see which materials explain VPC1
# The R² tells us how well they explain it
# Next we'll look at individual coefficients
# ============================================================
````

# PART 8: Understanding VIF (Variance Inflation Factor)
---

## 💡 What is VIF?

VIF measures **multicollinearity** - when predictor variables are too correlated with each other.

### Why It Matters:
- If two materials have similar spectra, they're **redundant**
- Using both causes **unstable coefficients** (can randomly flip sign!)
- Hard to tell which material is really contributing

### The Formula:
```
VIF = 1 / (1 - R²)
```
Where R² comes from regressing one predictor against ALL other predictors.

### VIF Values:
| VIF | Meaning | Action |
|-----|---------|--------|
| **= 1.0** | No correlation | Perfect |
| **< 2.0** (less than 2.0) | Low correlation | Good for VPCA |
| **< 5.0** (less than 5.0) | Moderate | OK for regular data |
| **> 10** (greater than 10) | High correlation | Remove variable! |

### Dr. Ortiz's Recommendation:
Use **VIF threshold = 2.0** for VPCA data (stricter than default 5.0)

### Why 2.0 and not 5.0?
VPCA components are **already partitioned** (orthogonalized). They should have low cross-correlation. A VIF of 2 means the variance is doubled due to collinearity - already too much for partitioned data.

---

## 🎯 Joint Stopping Criterion (The Key Innovation!)

Our stepwise regression uses **BOTH** criteria together:
1. **P-value < 0.05** (less than 0.05) = statistically significant
2. **VIF < 2.0** (less than 2.0) = no multicollinearity

A variable must pass **BOTH** to be included!

This is better than SPSS which only checks p-values.

---

````python
# ============================================================
# DEFINE VIF CALCULATION FUNCTION
# ============================================================

# LINE 1: Define the function
# 'X' is a DataFrame of predictor variables (NO constant column!)
def calculate_vif(X):
    """
    Calculate Variance Inflation Factor for each variable.
    
    VIF = 1 / (1 - R²)
    where R² is from regressing each variable against all others.
    
    Parameters:
        X: DataFrame of predictor variables
    
    Returns:
        DataFrame with 'Variable' and 'VIF' columns
    """
    
    # LINE 2: Create empty list to store results
    vif_data = []
    
    # LINE 3: Loop through each column (variable)
    for col in X.columns:
        
        # LINE 4: If only 1 variable, VIF is always 1.0
        # (nothing to be correlated with!)
        if len(X.columns) < 2:
            vif_data.append({'Variable': col, 'VIF': 1.0})
            continue
        
        # LINE 5: Get all OTHER columns (everything except this one)
        # List comprehension: [c for c in X.columns if c != col]
        # This creates a list of all column names EXCEPT 'col'
        other_cols = [c for c in X.columns if c != col]
        
        # LINE 6: Set up mini-regression
        # Y = this variable
        # X = all other variables (with constant added)
        y_vif = X[col]
        X_vif = sm.add_constant(X[other_cols])
        
        # LINE 7: Fit the model and calculate VIF
        try:
            model = sm.OLS(y_vif, X_vif).fit()
            r2 = model.rsquared
            
            # LINE 8: Apply VIF formula
            # If R² is close to 1.0 → VIF becomes huge (bad!)
            # If R² is close to 0.0 → VIF is close to 1.0 (good!)
            if r2 < 1.0:
                vif = 1.0 / (1.0 - r2)
            else:
                vif = float('inf')  # Infinity = perfect collinearity
        except:
            vif = float('inf')
        
        # LINE 9: Store result
        vif_data.append({'Variable': col, 'VIF': round(vif, 2)})
    
    # LINE 10: Convert list to DataFrame and return
    return pd.DataFrame(vif_data)

# ============================================================
# FUNCTION DEFINED!
# Usage: vif_df = calculate_vif(your_dataframe)
# ============================================================

print("✓ calculate_vif() function defined!")
print("  Usage: vif_df = calculate_vif(X)")
````

# PART 9: Stepwise Selection with VIF Control
---

## 💡 Three Methods of Stepwise Selection

### 1. Forward (start empty, ADD variables)
```
1. Start with NO materials
2. Try adding each remaining material
3. Check: p < (less than) 0.05 AND VIF < (less than) 2.0?
4. If BOTH pass → ADD it
5. If p >= (greater than or equal) 0.05 → STOP
```

### 2. Backward (start full, REMOVE variables)
```
1. Start with ALL materials
2. Remove any with VIF > (greater than) 2.0
3. Find material with highest p-value
4. If p > (greater than) 0.10 → REMOVE it
5. If all p <= (less than or equal) 0.10 → STOP
```

### 3. Bidirectional (BEST - combines both!)
```
1. Start empty
2. FORWARD: Try adding best material (p < 0.05 AND VIF < 2.0)
3. BACKWARD: Check if any should be removed (p > 0.10)
4. If no changes → STOP (converged)
```

### Why Bidirectional is Best:
- Forward can miss removing variables that become redundant later
- Backward can remove variables too early before seeing combinations
- Bidirectional **self-corrects** at every step!

### Our Settings (Per Professor):
| Parameter | Value | Meaning |
|-----------|-------|---------|
| `p_enter` | **0.05** | Add only if p < (less than) 0.05 |
| `p_remove` | **0.10** | Remove if p > (greater than) 0.10 |
| `vif_threshold` | **2.0** | All VIF must be < (less than) 2.0 |

---

````python
# ============================================================
# COMPLETE STEPWISE FUNCTION WITH VIF CONTROL
# Supports: forward, backward, bidirectional
# Uses JOINT criterion: p-value + VIF
# ============================================================

def stepwise_with_vif(y, X, method='bidirectional', p_enter=0.05, p_remove=0.10,
                      vif_threshold=2.0, verbose=True):
    """
    Stepwise regression with joint p-value + VIF stopping criterion.
    
    Methods:
        'forward': Start empty, add variables (check p + VIF)
        'backward': Start full, remove by VIF then p-value
        'bidirectional': Forward + backward each iteration (BEST)
    
    Parameters:
        y: Target variable (VPC loadings)
        X: Predictor variables (spectral library)
        method: 'forward', 'backward', or 'bidirectional'
        p_enter: Add if p < (less than) this (default 0.05)
        p_remove: Remove if p > (greater than) this (default 0.10)
        vif_threshold: Max VIF allowed, must be < (less than) this (default 2.0)
        verbose: Print step-by-step progress
    
    Returns:
        selected: List of selected variable names
        history: List of step details (dictionaries)
    """
    
    # Initialize based on method
    if method == 'backward':
        selected = list(X.columns)  # Start with ALL variables
        remaining = []
    else:  # forward or bidirectional
        selected = []  # Start empty
        remaining = list(X.columns)
    
    history = []
    step = 0
    
    if verbose:
        print(f"Stepwise ({method.upper()}) | p_enter={p_enter}, "
              f"p_remove={p_remove}, VIF_threshold={vif_threshold}")
        print("=" * 70)
    
    # ============================================================
    # BACKWARD: First remove high-VIF variables
    # ============================================================
    if method == 'backward':
        while len(selected) > 1:
            vif_df = calculate_vif(X[selected])
            worst = vif_df.loc[vif_df['VIF'].idxmax()]
            if worst['VIF'] <= vif_threshold:
                break  # All VIFs are OK
            step += 1
            selected.remove(worst['Variable'])
            remaining.append(worst['Variable'])
            history.append({
                'Step': step, 'Action': 'REMOVE(VIF)',
                'Variable': worst['Variable'],
                'P_value': None, 'R_squared': None,
                'VIF': round(worst['VIF'], 2)
            })
            if verbose:
                print(f"Step {step}: - {worst['Variable']:<15} "
                      f"(VIF={worst['VIF']:.2f} > (greater than) {vif_threshold}) REMOVED")
    
    # ============================================================
    # MAIN LOOP
    # ============================================================
    for _ in range(100):  # max iterations safety limit
        made_change = False
        
        # ==================== FORWARD STEP ====================
        # Try adding the best variable that passes BOTH criteria
        if method in ['forward', 'bidirectional'] and remaining:
            best_p, best_var, best_r2, best_coef = 1.0, None, 0.0, 0.0
            
            # Try each remaining variable
            for var in remaining:
                test_vars = selected + [var]
                try:
                    m = sm.OLS(y, sm.add_constant(X[test_vars])).fit()
                    if m.pvalues[var] < best_p:
                        best_p = m.pvalues[var]
                        best_var = var
                        best_r2 = m.rsquared
                        best_coef = m.params[var]
                except:
                    continue
            
            # Check JOINT criterion: p-value AND VIF must BOTH pass
            if best_var is not None and best_p < p_enter:
                test_vars = selected + [best_var]
                if len(test_vars) > 1:
                    vif_df = calculate_vif(X[test_vars])
                    max_vif = vif_df['VIF'].max()
                else:
                    max_vif = 1.0  # Single variable always has VIF = 1.0
                
                if max_vif <= vif_threshold:
                    # PASSES BOTH criteria → ADD variable
                    step += 1
                    selected.append(best_var)
                    remaining.remove(best_var)
                    made_change = True
                    history.append({
                        'Step': step, 'Action': 'ADD',
                        'Variable': best_var,
                        'P_value': round(best_p, 4),
                        'R_squared': round(best_r2, 4),
                        'VIF': round(max_vif, 2)
                    })
                    if verbose:
                        print(f"Step {step}: + ADD    {best_var:<15} "
                              f"| p={best_p:.4f} | R²={best_r2:.4f} | VIF={max_vif:.2f}")
                else:
                    # VIF too high → skip this variable permanently
                    if verbose:
                        print(f"  SKIP  {best_var:<15} "
                              f"| p={best_p:.4f} OK but VIF={max_vif:.2f} "
                              f"> (greater than) {vif_threshold} TOO HIGH")
                    remaining.remove(best_var)
            
            elif method == 'forward' and not made_change:
                if verbose:
                    if best_var:
                        print(f"\nStopping: best p={best_p:.4f} "
                              f">= (greater than or equal) {p_enter}")
                    else:
                        print("\nStopping: no variables to add")
                break
        
        # ==================== BACKWARD STEP ====================
        # Check if any current variable should be removed
        if method in ['backward', 'bidirectional'] and len(selected) > 0:
            try:
                m = sm.OLS(y, sm.add_constant(X[selected])).fit()
                pvals = m.pvalues.drop('const', errors='ignore')
                
                if not pvals.empty:
                    worst_var = pvals.idxmax()
                    worst_p = pvals.max()
                    
                    if worst_p > p_remove:
                        # P-value too high → REMOVE
                        step += 1
                        selected.remove(worst_var)
                        remaining.append(worst_var)
                        made_change = True
                        new_r2 = (sm.OLS(y, sm.add_constant(X[selected])).fit().rsquared
                                  if selected else 0.0)
                        max_vif = (calculate_vif(X[selected])['VIF'].max()
                                   if len(selected) > 1 else
                                   (1.0 if selected else 0.0))
                        history.append({
                            'Step': step, 'Action': 'REMOVE',
                            'Variable': worst_var,
                            'P_value': round(worst_p, 4),
                            'R_squared': round(new_r2, 4),
                            'VIF': round(max_vif, 2)
                        })
                        if verbose:
                            print(f"Step {step}: - REMOVE {worst_var:<15} "
                                  f"| p={worst_p:.4f} > (greater than) {p_remove}")
            except:
                pass
            
            # Also check VIF and remove if too high
            if len(selected) > 1:
                vif_df = calculate_vif(X[selected])
                worst = vif_df.loc[vif_df['VIF'].idxmax()]
                if worst['VIF'] > vif_threshold:
                    step += 1
                    selected.remove(worst['Variable'])
                    remaining.append(worst['Variable'])
                    made_change = True
                    history.append({
                        'Step': step, 'Action': 'REMOVE(VIF)',
                        'Variable': worst['Variable'],
                        'P_value': None, 'R_squared': None,
                        'VIF': round(worst['VIF'], 2)
                    })
                    if verbose:
                        print(f"Step {step}: - REMOVE {worst['Variable']:<15} "
                              f"| VIF={worst['VIF']:.2f} > (greater than) {vif_threshold}")
        
        # Check convergence
        if not made_change:
            if verbose:
                if method == 'bidirectional':
                    print(f"\nConverged: no more changes needed")
                elif method == 'backward':
                    print(f"\nStopping: all p-values <= (less than or equal) {p_remove}")
            break
    
    # ============================================================
    # PRINT FINAL SUMMARY WITH VIF FOR EVERY SELECTED MATERIAL
    # ============================================================
    if verbose:
        print("=" * 70)
        if selected:
            print(f"RESULT: Selected {len(selected)} materials: {selected}")
            print()
            # Always show VIF for every selected material
            if len(selected) > 1:
                vif_df = calculate_vif(X[selected])
            else:
                # Single variable: VIF is always 1.0
                vif_df = pd.DataFrame([{'Variable': selected[0], 'VIF': 1.0}])
            print(f"  {'Material':<20} {'VIF':>8}  {'Status'}")
            print(f"  {'-'*20} {'-'*8}  {'-'*20}")
            for _, row in vif_df.iterrows():
                status = "✓ OK (< 2.0)" if row['VIF'] < vif_threshold else "✗ TOO HIGH"
                print(f"  {row['Variable']:<20} {row['VIF']:8.2f}  {status}")
        else:
            print("RESULT: No materials selected")
    
    return selected, history

# ============================================================
print("✓ stepwise_with_vif() function defined!")
print("  Method: bidirectional (default, best method)")
print("  Joint criterion: p < (less than) 0.05 AND VIF < (less than) 2.0")
````

````python
---
# PART 10: Analyze All VPCs (Bidirectional with VIF)
---

Now we run **bidirectional stepwise with VIF control** for **every VPC** automatically!

### Settings (Per Professor):
| Parameter | Value | Meaning |
|-----------|-------|---------|
| **Method** | `bidirectional` | Best method - combines forward + backward |
| **p_enter** | `0.05` | Add only if p < (less than) 0.05 |
| **p_remove** | `0.10` | Remove if p > (greater than) 0.10 |
| **VIF threshold** | `2.0` | All VIF must be < (less than) 2.0 |

### What Happens:
- The function loops through VPC1, VPC2, VPC3, VPC4, VPC5, VPC6
- For each VPC, it runs bidirectional stepwise with VIF control
- It stores the results including **R²**, **p-values**, and **VIF** for every material
- At the end, it displays a summary table and saves to CSV

---
````

````python
# ============================================================
# RUN BIDIRECTIONAL STEPWISE WITH VIF FOR ALL VPCs
# ============================================================

# Settings (per Professor's methodology)
P_ENTER = 0.05          # Add only if p < (less than) 0.05
P_REMOVE = 0.10         # Remove if p > (greater than) 0.10
VIF_THRESHOLD = 2.0     # All VIF must be < (less than) 2.0
METHOD = 'bidirectional'  # Best method (combines forward + backward)

print(f"Analyzing All VPCs: {METHOD.upper()} with VIF Control")
print(f"Settings: p_enter={P_ENTER}, p_remove={P_REMOVE}, VIF < (less than) {VIF_THRESHOLD}")
print("=" * 70)

# Create empty list to store results for ALL VPCs
all_results = []

# Loop through each VPC column (VPC1, VPC2, ..., VPC6)
for vpc in z_loadings.columns:
    
    print(f"\n{'='*70}")
    print(f"Analyzing {vpc}")
    print(f"{'='*70}")
    
    # Run bidirectional stepwise with VIF control
    selected, history = stepwise_with_vif(
        y=z_loadings[vpc],       # Target: this VPC's loadings
        X=z_library,             # Predictors: all materials
        method=METHOD,           # Bidirectional (best)
        p_enter=P_ENTER,         # 0.05 threshold
        p_remove=P_REMOVE,       # 0.10 threshold
        vif_threshold=VIF_THRESHOLD,  # 2.0 for VPCA
        verbose=True             # Show step-by-step output
    )
    
    # Fit final model and get VIF details for selected materials
    if selected:
        X_final = sm.add_constant(z_library[selected])
        model = sm.OLS(z_loadings[vpc], X_final).fit()
        r2 = model.rsquared
        adj_r2 = model.rsquared_adj
        
        # Calculate VIF for each selected material
        if len(selected) > 1:
            vif_df = calculate_vif(z_library[selected])
            max_vif = vif_df['VIF'].max()
            # Build VIF detail string: "Water(1.23), Clay(1.45)"
            vif_details = ', '.join([f"{row['Variable']}({row['VIF']:.2f})" 
                                     for _, row in vif_df.iterrows()])
        else:
            max_vif = 1.0
            vif_details = f"{selected[0]}(1.00)"
    else:
        r2 = 0.0
        adj_r2 = 0.0
        max_vif = 0.0
        vif_details = 'N/A'
    
    # Store results for this VPC
    all_results.append({
        'Component': vpc,
        'N_Materials': len(selected),
        'Materials': ', '.join(selected) if selected else 'None',
        'R_squared': round(r2, 4),
        'Adj_R_squared': round(adj_r2, 4),
        'Max_VIF': round(max_vif, 2),
        'VIF_Details': vif_details
    })

print("\n" + "="*70)
print("✓ Analysis Complete for All VPCs!")
print("="*70)
````
